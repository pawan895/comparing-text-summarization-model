```python
import json
import matplotlib.pyplot as plt

# Function to read ROUGE scores from file
def read_rouge_scores(file_path):
    with open(file_path, 'r') as file:
        scores = json.load(file)
    return scores

# File paths for ROUGE scores
file_paths = {
    'Pagus': 'results/pagus_rouge_scores.txt',
    'LexRank': 'results/lexrank_rouge_scores.txt',
    'Hugging Face': 'results/hugging_face_rouge_scores.txt'
}

# Initialize lists to store ROUGE scores
rouge_1_r = []
rouge_1_p = []
rouge_1_f = []

rouge_2_r = []
rouge_2_p = []
rouge_2_f = []

rouge_l_r = []
rouge_l_p = []
rouge_l_f = []

# Read ROUGE scores from files
for model, file_path in file_paths.items():
    scores = read_rouge_scores(file_path)
    
    # Extract ROUGE scores
    rouge_1_r.append(scores[0]['rouge-1']['r'])
    rouge_1_p.append(scores[0]['rouge-1']['p'])
    rouge_1_f.append(scores[0]['rouge-1']['f'])
    
    rouge_2_r.append(scores[0]['rouge-2']['r'])
    rouge_2_p.append(scores[0]['rouge-2']['p'])
    rouge_2_f.append(scores[0]['rouge-2']['f'])
    
    rouge_l_r.append(scores[0]['rouge-l']['r'])
    rouge_l_p.append(scores[0]['rouge-l']['p'])
    rouge_l_f.append(scores[0]['rouge-l']['f'])

# ROUGE scores for each model
models = list(file_paths.keys())

# Plotting ROUGE-1 Recall, Precision, and F1 scores
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.bar(models, rouge_1_r, color='skyblue')
plt.ylabel('ROUGE-1 Recall')
plt.title('ROUGE-1 Recall')

plt.subplot(1, 3, 2)
plt.bar(models, rouge_1_p, color='salmon')
plt.ylabel('ROUGE-1 Precision')
plt.title('ROUGE-1 Precision')

plt.subplot(1, 3, 3)
plt.bar(models, rouge_1_f, color='lightgreen')
plt.ylabel('ROUGE-1 F1 Score')
plt.title('ROUGE-1 F1 Score')

plt.tight_layout()
plt.show()

# Plotting ROUGE-2 Recall, Precision, and F1 scores
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.bar(models, rouge_2_r, color='skyblue')
plt.ylabel('ROUGE-2 Recall')
plt.title('ROUGE-2 Recall')

plt.subplot(1, 3, 2)
plt.bar(models, rouge_2_p, color='salmon')
plt.ylabel('ROUGE-2 Precision')
plt.title('ROUGE-2 Precision')

plt.subplot(1, 3, 3)
plt.bar(models, rouge_2_f, color='lightgreen')
plt.ylabel('ROUGE-2 F1 Score')
plt.title('ROUGE-2 F1 Score')

plt.tight_layout()
plt.show()

# Plotting ROUGE-L Recall, Precision, and F1 scores
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.bar(models, rouge_l_r, color='skyblue')
plt.ylabel('ROUGE-L Recall')
plt.title('ROUGE-L Recall')

plt.subplot(1, 3, 2)
plt.bar(models, rouge_l_p, color='salmon')
plt.ylabel('ROUGE-L Precision')
plt.title('ROUGE-L Precision')

plt.subplot(1, 3, 3)
plt.bar(models, rouge_l_f, color='lightgreen')
plt.ylabel('ROUGE-L F1 Score')
plt.title('ROUGE-L F1 Score')

plt.tight_layout()
plt.show()
```

```python
# Calculate average F1 score for each model
average_f1_scores = {
    model: (sum([rouge_1_f[i], rouge_2_f[i], rouge_l_f[i]]) / 3)
    for i, model in enumerate(models)
}

# Find the model with the highest average F1 score
best_model = max(average_f1_scores, key=average_f1_scores.get)

print("Best model based on average F1 score:", best_model)
```
